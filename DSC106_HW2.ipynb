```python
daily_sales_table['HM_total'] = daily_sales_table['HM-NE'] + daily_sales_table['HM-SW'] + daily_sales_table['HM-NW'] + daily_sales_table['HM-SE'] + daily_sales_table['HM-C']
daily_sales_table['CF_total'] = daily_sales_table['CF-NE'] + daily_sales_table['CF-SW'] + daily_sales_table['CF-NW'] + daily_sales_table['CF-SE'] + daily_sales_table['CF-C']
daily_sales_table['FF_total'] = daily_sales_table['FF-NE'] + daily_sales_table['FF-SW'] + daily_sales_table['FF-NW'] + daily_sales_table['FF-SE'] + daily_sales_table['FF-C']
```

```python
daily_sales_table
```

```python
new_daily_sales_table = daily_sales_table.drop('Day', axis=1)
```

```python
new_daily_sales_table = daily_sales_table[['Unnamed: 0', 'HM_total', 'CF_total', 'FF_total']]
```

```python
new_daily_sales_table
```

```python
new_daily_sales_table = new_daily_sales_table.groupby('Unnamed: 0').mean()
```

```python
new_daily_sales_table['Day of week'] = new_daily_sales_table.index
```
